```python
import pandas as pd
```

```python
nomes_colunas = [
    'Status_Conta_Corrente', 'Duracao_Mes', 'Historico_Credito', 'Proposito',
    'Valor_Credito', 'Conta_Poupanca', 'Tempo_Emprego_Atual', 'Taxa_Parcelamento',
    'Estado_Civil_Sexo', 'Outros_Devedores', 'Residencia_Atual_Desde', 'Propriedade',
    'Idade', 'Outros_Planos_Pagamento', 'Moradia', 'N_Creditos_Existentes',
    'Trabalho', 'N_Pessoas_Sustento', 'Telefone', 'Trabalhador_Estrangeiro', 'Risco']
```

```python
df = pd.read_csv('german.data',
                 sep=' ',
                 header=None,
                 names=nomes_colunas
                 )
```

```python
df.info()
```

```python
df.head()
```

```python
#Criando mapas (Traduções):

mapa_status_conta = {
    'A11': '< 0 DM',
    'A12': '0 a 200 DM',
    'A13': '>= 200 DM',
    'A14': 'Sem conta'
}

mapa_proposito = {
    'A40': 'Carro Novo', 'A41': 'Carro Usado', 'A42': 'Móveis/Equipamento',
    'A43': 'Rádio/TV', 'A44': 'Eletrodomésticos', 'A45': 'Reparos',
    'A46': 'Educação', 'A47': 'Férias', 'A48': 'Retreinamento',
    'A49': 'Negócios', 'A410': 'Outros'
}

mapa_poupanca = {
    'A61': '< 100 DM', 'A62': '100 a 500 DM', 'A63': '500 a 1000 DM',
    'A64': '>= 1000 DM', 'A65': 'Sem conta'
}

mapa_risco = {1: 'Bom',
              2: 'Ruim'}

mapa_tempo_emprego = {
    'A71': 'Desempregado',
    'A72': '< 1 ano',
    'A73': '1  a 4 anos',
    'A74': '4 a 7 anos',
    'A75': '>= 7 anos'
}

mapa_moradia = {
    'A151': 'Aluguel',
    'A152': 'Própria',
    'A153': 'Favor'
}

mapa_ocupacao = {
    'A171': 'Desempregado',
    'A172': 'Informal',
    'A173': 'Qualificado',
    'A174': 'Autônomo/Gerência/Oficial'
}

mapa_telefone = {
    'A191': 'Sim',
    'A192': 'Não'
}

mapa_estrangeiro = {
    'A201': 'Sim',
    'A202': 'Não'
}

df.head()
```

```python
#Aplicando a tradução:

df['Status_Conta_Corrente'] = df['Status_Conta_Corrente'].replace(mapa_status_conta)
df['Proposito'] = df['Proposito'].replace(mapa_proposito)
df['Conta_Poupanca'] = df['Conta_Poupanca'].replace(mapa_poupanca)
df['Risco'] = df['Risco'].replace(mapa_risco)
df['Tempo_Emprego_Atual'] = df['Tempo_Emprego_Atual'].replace(mapa_tempo_emprego)
df['Moradia'] = df['Moradia'].replace(mapa_moradia)
df['Trabalho'] = df['Trabalho'].replace(mapa_ocupacao)
df['Telefone'] = df['Telefone'].replace(mapa_telefone)
df['Trabalhador_Estrangeiro'] = df['Trabalhador_Estrangeiro'].replace(mapa_estrangeiro)
```

```python
df.head()
```

```python
#Verificando dados faltantes
print(df.isnull().sum())
```

```python
#Definindo uma nova coluna (Faixa Etária) para melhor análise (Engenharia de atributos)

bins_idade = [18, 30, 40, 50, 60, 100]
labels_idade = ['18-30 anos', '31-40 anos', '41-50 anos', '51-60 anos', '60+ anos']

#Criando:
#Essa linha cria uma nova coluna que será a idade separado por bins
df['Faixa_Etaria'] = pd.cut(df['Idade'], bins=bins_idade, labels=labels_idade)

df[['Idade','Faixa_Etaria']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 7)

df['Risco'].value_counts()
```

```python
sns.countplot(x='Risco', data=df)

plt.title('Distribuição de Clientes por Risco')
plt.xlabel('Classificação de Risco')
plt.ylabel('Quantidade de Clientes')
plt.show()
```

Com o gráfico abaixo é possível notar que gastos com educação tende a ter um maior risco.

```python
sns.countplot(y='Proposito', hue='Risco', data=df , order=df['Proposito'].value_counts().index)

plt.title('Análise de Risco por Propósito do Crédito')
plt.xlabel('Quantidade de Clientes')
plt.ylabel('Propósito do Crédito')
plt.show()
```

Com o gráfico abaixo é possível notar que pessoas mais novas tendem a ser um mal pagador em relação as outras idades

```python
sns.countplot(y='Faixa_Etaria', hue='Risco', data=df,)
plt.title('Análise de Risco por Faixa Etária')
plt.xlabel('Quantidade de Clientes')
plt.ylabel('Faixa Etária')
plt.show()
```

Ao observar o boxplot abaixo é possível notar que o risco está ligado ao valor de crédito ofertado, quanto maior o crédito maior a chance de inadimplência

```python
sns.boxplot(x='Risco', y='Valor_Credito', data=df)
plt.title('Distribuição do Valor do Crédito por Risco')
plt.xlabel('Classificação de Risco')
plt.ylabel('Valor do Crédito (em DM)')
plt.show()
```

```python
sns.countplot(x='Moradia', hue='Risco', data=df)

plt.title('Análise de Risco por Moradia')
plt.xlabel('Moradia')
plt.ylabel('Quantidade de Clientes')
plt.show()
```
